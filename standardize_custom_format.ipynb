```python
def standardize_custom_format(conversation):
  parts = conversation.split("'answer': ")
  if "'question': " in parts[0]:
      question_part = parts[0].split("'question': ")[1]
  else:
    question_part = ""
  answer_part = parts[1]
  return {
        "messages": [
            {"role": "user", "content": question_part},
            {"role": "assistant", "content": answer_part},
        ]
    }
```
