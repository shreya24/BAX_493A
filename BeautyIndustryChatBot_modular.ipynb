Next, I calculate answer relevancy score for questions asking ***about ingredients used in the beauty products.***   
Source for understanding relevance score- https://docs.ragas.io/en/v0.1.21/concepts/metrics/answer_relevance.html.  
The evaluation metric, Answer Relevancy, focuses on assessing how pertinent the generated answer is to the given prompt. A lower score is assigned to answers that are incomplete or contain redundant information and higher scores indicate better relevancy.    
AnswerRelevance.score_batch() takes two parallel lists: the input questions (prompts) and the corresponding model-generated responses, and returns a relevance score between 0 and 1 for each pair   
I am using this metric as I do not have ground truth answers for questions asking about safety of ingredients. I am unable to calculate the ROUGE score for this reason. There is no such dataset available. Answer Relevance can be calulated without ground truth answers or context.  
  
There is this difficulty in calculating a base matric. But this is what makes this project exciting to me. Lack of such a curated dataset suggests that no such work on a large scale exists today.

```python
def generate_response(prompt):
    """Simple function to generate response from model"""
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(
        **inputs,
        max_new_tokens=100,
        do_sample=False,
        pad_token_id=tokenizer.eos_token_id
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # Extract only the new generated part
    return response[len(prompt):].strip()
```

Using prompt-chaining with Qwen/Qwen2.5-1.5B-Instruct below.   
First LLM generates topics basis reviews in the current batch. Second LLM removes topics which were already covered in previous iterations.

```python
previous_topics_broadcast = spark.sparkContext.broadcast(set())
TEXT_DELIMITER = '####'

for i, batch in enumerate(partitions, start=1):
    print(f"\nProcessing Batch {i}")

    previous_topics = previous_topics_broadcast.value

    # First LLM call: Generate 5 summarized topics
    prompt_1 = f"""Below is a set of reviews delimited by {TEXT_DELIMITER}:\n{batch}\n
Identify 5 summarized topics (1‚Äì4 words) from these reviews.
Return the output as a bullet list, each line one concise topic. No descriptions."""

    text_1 = generate_response(prompt_1)
    print("Raw Topics Output:", text_1)

    extracted_topics = []
    for line in text_1.split("\n"):
        if line.strip():
            match = re.search(r"[-‚Ä¢*]?\s*(.+)", line.strip())
            if match:
                extracted_topics.append(match.group(1).strip())

    print("Extracted Topics:", extracted_topics)

    # Second LLM call: Filter out previously seen topics
    prompt_2 = f"""Given the following topics:\n{extracted_topics}\n
Previously covered topics are: {list(previous_topics)}\n
Return only the new topics not present in the previously covered list.
Output as a bullet list, one topic per line, 1‚Äì4 words, no description."""

    text_2 = generate_response(prompt_2)
    print("Filtered Topics Output:", text_2)

    new_topics = []
    for line in text_2.split("\n"):
        if line.strip():
            match = re.search(r"[-‚Ä¢*]?\s*(.+)", line.strip())
            if match:
                topic = match.group(1).strip()
                if topic:
                    new_topics.append(topic)

    print("New Topics:", new_topics)

    # Update broadcasted topics
    if new_topics:
        updated_set = previous_topics.union(set(new_topics))
        previous_topics_broadcast.unpersist()
        previous_topics_broadcast = spark.sparkContext.broadcast(updated_set)

    print("Updated Unique Topics:", previous_topics_broadcast.value)
    print("-" * 50)

    if i == 10: break

print("Final Unique Topics:", previous_topics_broadcast.value)
```
