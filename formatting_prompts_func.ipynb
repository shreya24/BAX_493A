```python
def formatting_prompts_func(examples, STANDARD_SYSTEM_PROMPT, tokenizer):
    print("In formatting_prompts_func")
    convos = examples["standardized"]
    new_texts = []

    for convo in convos:
        messages = convo["messages"]

        # Insert custom system prompt at the start
        if not (messages and messages[0]["role"] == "system"):
            messages = [{"role": "system", "content": STANDARD_SYSTEM_PROMPT}] + messages
        else:
            messages[0]["content"] = STANDARD_SYSTEM_PROMPT

        # Apply chat template
        text = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=False
        )
        new_texts.append(text)

    print(new_texts)
    return { "text": new_texts }
```
